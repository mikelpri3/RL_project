```python
import pandas as pd
from data import generar_csv_limpio

# Cargamos y limpiamos el DataFrame usando la función de data.py
pokemon_clean = generar_csv_limpio()

pokemon_clean
```

```python
# Buscamos en la columna 'id' el valor 123
pokemon_id_123 = pokemon_clean[pokemon_clean['id'] == 123]

print("Pokémon con ID 123:")
print(pokemon_id_123)
```

```python
# Buscamos en la columna 'name' el string 'Victini'
victini = pokemon_clean[pokemon_clean['name'] == 'Victini']

print("Pokémon con el nombre Victini:")
print(victini)
```

```python
# Filtramos buscando 'Dragon' en la columna 'type1' O en la columna 'type2'
# Usamos .head(10) para limitar el resultado a los primeros 10 que encuentre
pokemons_dragon = pokemon_clean[
    (pokemon_clean['type1'] == 'Dragon') | (pokemon_clean['type2'] == 'Dragon')
].head(10)


print("10 Pokémon de tipo Dragón:")
print(pokemons_dragon)
```

```python
# Usamos idxmax() para encontrar el índice de la fila con el valor máximo en la columna 'Speed'
pokemon_mas_rapido = pokemon_clean.loc[pokemon_clean['Speed'].idxmax()]

# Hacemos lo mismo con idxmin() para el valor mínimo
pokemon_mas_lento = pokemon_clean.loc[pokemon_clean['Speed'].idxmin()]


print("--- Pokémon con MÁS velocidad ---")
print(pokemon_mas_rapido)
print("\n" + "="*40 + "\n")
print("--- Pokémon con MENOS velocidad ---")
print(pokemon_mas_lento)
```

```python
# Importamos la función necesaria de nuestro archivo data.py
from data import crear_todos_los_movimientos
import itertools

# Ejecutamos la función para obtener el diccionario de movimientos
diccionario_de_movimientos = crear_todos_los_movimientos()

print(f"Se han creado un total de {len(diccionario_de_movimientos)} movimientos.")
print("\n--- Mostrando los primeros 5 movimientos del diccionario: ---")

# Un diccionario no tiene un método .head() como un DataFrame.
# Usamos itertools.islice para obtener los primeros 5 elementos de forma eficiente.
primeros_cinco = dict(itertools.islice(diccionario_de_movimientos.items(), 5))

# Imprimimos cada uno de los 5 movimientos
for nombre, objeto_movimiento in primeros_cinco.items():
    print(objeto_movimiento)
```

```python

# Importamos la nueva función y las librerías para mostrar imágenes
from data import crear_todos_los_entrenadores
from IPython.display import Image, display

# Ejecutamos la función para obtener la lista de entrenadores
lista_de_entrenadores = crear_todos_los_entrenadores()

# Verificamos si la lista no está vacía y seleccionamos el primer entrenador
if lista_de_entrenadores:
    primer_entrenador = lista_de_entrenadores[0]

    print("--- Datos del primer Entrenador de la lista ---")
    print(f"ID:     {primer_entrenador.id}")
    print(f"Nombre: {primer_entrenador.name}")
    print(f"Ruta Img: {primer_entrenador.imagen}")
    
    print("\n--- Equipo Pokémon ---")
    for pokemon in primer_entrenador.pokemons:
        # Imprimimos el nombre y los tipos de cada Pokémon en su equipo
        tipos_str = f"{pokemon.type1}"
        if pokemon.type2:
            tipos_str += f" / {pokemon.type2}"
        print(f"- {pokemon.name} (Tipos: {tipos_str})")

    # --- Mostramos la imagen del entrenador ---
    print("\n--- Imagen del Entrenador ---")
    try:
        imagen_entrenador = Image(filename=primer_entrenador.imagen)
        display(imagen_entrenador)
    except FileNotFoundError:
        print(f"No se encontró la imagen en la ruta: {primer_entrenador.imagen}")

else:
    print("No se ha creado ningún entrenador. Revisa el archivo trainer_pokemon.csv.")
```

```python
# Importamos las funciones y librerías necesarias
from image_generator import crear_imagen_combate
from data import crear_todos_los_entrenadores
from IPython.display import Image, display

print("Cargando datos de entrenadores para la simulación...")
entrenadores = crear_todos_los_entrenadores()

if entrenadores:
    # Ambos entrenadores son el mismo, 'trainer1'
    entrenador_principal = entrenadores[0]

    # Buscamos los Pokémon específicos por su nombre para mayor seguridad
    pokemon_jugador = next((p for p in entrenador_principal.pokemons if p.name == "Blastoise"), None)
    pokemon_oponente = next((p for p in entrenador_principal.pokemons if p.name == "Venusaur"), None)

    # Solo procedemos si ambos Pokémon fueron encontrados
    if pokemon_jugador and pokemon_oponente:
        # Definimos el texto que aparecerá en el panel de la derecha
        texto_del_combate = (
            "------ TURNO 1 ------\n\n"
            "Blastoise utiliza Hidropulso,\n"
            "hace 20.62% de daño.\n"
            "Es poco efectivo...\n\n"
            "Venusaur utiliza Lluevehojas,\n"
            "hace 0% de daño.\n"
            "¡Blastoise esquivó el ataque!"
        )

        print(f"Preparando combate: {pokemon_jugador.name} vs. {pokemon_oponente.name}")

        # Definimos la ruta de salida y llamamos a la función
        ruta_imagen_generada = "combate_con_panel.png"
        crear_imagen_combate(
            entrenador1=entrenador_principal, 
            pokemon1=pokemon_jugador,
            entrenador2=entrenador_principal, 
            pokemon2=pokemon_oponente,
            texto_combate=texto_del_combate,
            output_path=ruta_imagen_generada
        )
        
        # Mostramos la imagen recién creada
        display(Image(filename=ruta_imagen_generada))
    else:
        print("Error: No se encontró a Blastoise o Venusaur en el equipo de 'trainer1'.")
else:
    print("Error: No se pudo cargar ningún entrenador desde el archivo CSV.")
```
