```python
# Importamos las funciones y librerías necesarias
from image_generator import crear_imagen_combate
from data import crear_todos_los_entrenadores
from IPython.display import Image, display

print("Cargando datos de entrenadores para la simulación...")
entrenadores = crear_todos_los_entrenadores()

if entrenadores:
    # Ambos entrenadores son el mismo, 'trainer1'
    entrenador_principal = entrenadores[0]

    # Buscamos los Pokémon específicos por su nombre para mayor seguridad
    pokemon_jugador = next((p for p in entrenador_principal.pokemons if p.name == "Blastoise"), None)
    pokemon_oponente = next((p for p in entrenador_principal.pokemons if p.name == "Venusaur"), None)

    # Solo procedemos si ambos Pokémon fueron encontrados
    if pokemon_jugador and pokemon_oponente:
        # Definimos el texto que aparecerá en el panel de la derecha
        texto_del_combate = (
            "------ TURNO 1 ------\n\n"
            "Blastoise utiliza Hidropulso,\n"
            "hace 20.62% de daño.\n"
            "Es poco efectivo...\n\n"
            "Venusaur utiliza Lluevehojas,\n"
            "hace 0% de daño.\n"
            "¡Blastoise esquivó el ataque!"
        )

        print(f"Preparando combate: {pokemon_jugador.name} vs. {pokemon_oponente.name}")

        # Definimos la ruta de salida y llamamos a la función
        ruta_imagen_generada = "combate_con_panel.png"
        crear_imagen_combate(
            entrenador1=entrenador_principal, 
            pokemon1=pokemon_jugador,
            entrenador2=entrenador_principal, 
            pokemon2=pokemon_oponente,
            texto_combate=texto_del_combate,
            output_path=ruta_imagen_generada
        )
        
        # Mostramos la imagen recién creada
        display(Image(filename=ruta_imagen_generada))
    else:
        print("Error: No se encontró a Blastoise o Venusaur en el equipo de 'trainer1'.")
else:
    print("Error: No se pudo cargar ningún entrenador desde el archivo CSV.")
```

```python
from collections import defaultdict
import numpy as np

n_actions = 4
Q = defaultdict(lambda: np.zeros(n_actions, dtype=np.float32))

s1, s2 = 10, 99
print("Before touching:", s1 in Q, s2 in Q)        # False False

# Access creates rows:
print("Q[s1] =", Q[s1])                            # [0. 0. 0. 0.]
Q[s1][2] = 1.5
print("After update, Q[s1] =", Q[s1])              # [0. 0. 1.5 0.]
print("Keys now:", list(Q.keys()))                 # [10]

_ = Q[s2]                                          # auto-creates row for s2
print("Keys now:", list(Q.keys()))                 # [10, 99]
```

smoothed training curve 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("logs/train_log_20251024_125727.csv")

# smoothing
win = 20
df["train_return_smooth"] = df["train_return"].rolling(window=win, min_periods=1).mean()

plt.figure(figsize=(9,5))
plt.plot(df["episode"], df["train_return"], alpha=0.25, label="raw train_return")
plt.plot(df["episode"], df["train_return_smooth"], label=f"smoothed ({win}-episode avg)")
plt.axhline(0, color="k", linewidth=0.8, alpha=0.4)
plt.xlabel("Episode")
plt.ylabel("Train return")
plt.title("Training return (raw & smoothed)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("logs/train_log_20251024_125727.csv")
eval_df = df[df["eval_winrate"].notna()].copy()

plt.figure(figsize=(9,5))
plt.plot(eval_df["episode"], eval_df["eval_winrate"], marker="o", linestyle="-", label="greedy win-rate")
plt.ylim(-0.05, 1.05)
plt.xlabel("Episode")
plt.ylabel("Win-rate (ε=0)")
plt.title("Learning curve (greedy evaluation)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Q-table growth & epsilon schedule

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("logs/train_log_20251024_125727.csv")

fig, ax1 = plt.subplots(figsize=(9,5))
ax1.plot(df["episode"], df["q_rows"], label="#Q rows (states)", alpha=0.9)
ax1.set_xlabel("Episode")
ax1.set_ylabel("#Q rows (unique states)")

ax2 = ax1.twinx()
ax2.plot(df["episode"], df["epsilon"], label="epsilon", color="tab:orange", alpha=0.8)
ax2.set_ylabel("epsilon")

lns = ax1.get_lines()+ax2.get_lines()
labs = [l.get_label() for l in lns]
plt.title("State coverage & exploration schedule")
plt.legend(lns, labs, loc="lower right")
plt.grid(True, axis="x")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("logs/train_log_20251024_125727.csv")
eval_df = df[df["eval_winrate"].notna()]

print("Episodes:", len(df))
print("Final epsilon:", round(df["epsilon"].iloc[-1], 4))
print("Q rows (min/mean/max):", int(df["q_rows"].min()), round(df["q_rows"].mean(),1), int(df["q_rows"].max()))
print("Train return (mean±std):", round(df["train_return"].mean(),2), "±", round(df["train_return"].std(),2))
print("Final train return:", round(df["train_return"].iloc[-1],2))
print("Eval points:", len(eval_df))
print("Final eval win-rate:", float(eval_df["eval_winrate"].iloc[-1]))
print("Eval returns (min/mean/max):",
      round(eval_df["eval_return"].min(),2),
      round(eval_df["eval_return"].mean(),2),
      round(eval_df["eval_return"].max(),2))
```
